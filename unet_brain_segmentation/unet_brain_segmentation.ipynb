```python
import os

# --- CONFIGURE YOUR PATHS HERE ---
# The main directory where the 'keras_png_slices...' folders are
base_data_dir = '/content/drive/MyDrive/colab_brains_gan/data'
# ---------------------------------

train_img_dir = os.path.join(base_data_dir, "keras_png_slices_train")
train_mask_dir = os.path.join(base_data_dir, "keras_png_slices_seg_train")

print("--- Checking Training Set ---")

# Get a set of all available mask filenames for fast checking
actual_mask_files = set(os.listdir(train_mask_dir))

# Get a list of all image files
image_files = [f for f in os.listdir(train_img_dir) if f.startswith('case_')]

orphan_images = []

for img_fn in image_files:
    # Generate the mask filename we expect to find
    expected_mask_fn = img_fn.replace("case_", "seg_")

    # Check if the expected mask file actually exists
    if expected_mask_fn not in actual_mask_files:
        orphan_images.append(img_fn)

if not orphan_images:
    print("✅ Training set looks consistent!")
else:
    print(f"Found {len(orphan_images)} orphan image(s) with no matching mask:")
    for orphan in orphan_images:
        print(f"  - {orphan}")

# You can repeat this process for your validation set by changing the directory paths.
```

```python
import os

# 1. VERIFY THIS IS THE CORRECT PATH to your training images
image_dir = '/content/drive/MyDrive/colab_brains_gan/data/keras_png_slices_train'

# 2. List of the exact orphan filenames to be deleted
files_to_delete = [
    "case_382_slice_21 (1).nii.png",
    "case_229_slice_6 (1).nii.png",
    "case_185_slice_20 (1).nii.png"
]

print("Starting deletion process...")

for filename in files_to_delete:
    # Construct the full path to the file
    full_path = os.path.join(image_dir, filename)

    # Check if the file exists before trying to delete it
    if os.path.exists(full_path):
        try:
            os.remove(full_path)
            print(f"🗑️ Successfully deleted: {filename}")
        except OSError as e:
            print(f"Error deleting {filename}: {e}")
    else:
        print(f"File not found, skipping: {filename}")

print("\nDeletion process complete.")
```
