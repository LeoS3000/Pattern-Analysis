```python
import torch
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.utils import save_image
import os
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt

# We can now import our custom modules
from model import VAE, vae_loss_function
from dataset import BrainScanDataset

# --- 1. Hyperparameters ---
DATA_DIR = '/content/drive/MyDrive/colab_brains_gan/data/keras_png_slices_train'
OUTPUT_DIR = 'outputs'
IMG_SIZE = 128
LATENT_DIM = 32
BATCH_SIZE = 1024
NUM_EPOCHS = 100
LEARNING_RATE = 1e-3

# --- 2. Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
os.makedirs(os.path.join(OUTPUT_DIR, "reconstructions"), exist_ok=True)

# --- 3. Data Loading ---
transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(), # Normalizes to [0, 1]
])
dataset = BrainScanDataset(root_dir=DATA_DIR, transform=transform)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)

# --- 4. Model, Optimizer, Loss ---
model = VAE(img_channels=1, latent_dim=LATENT_DIM).to(device)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
loss_history = []

# --- 5. Training Loop ---
print("Starting training...")
for epoch in range(NUM_EPOCHS):
    model.train()
    total_loss = 0
    progress_bar = tqdm(dataloader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}")

    for batch in progress_bar:
        images = batch.to(device)
        optimizer.zero_grad()
        recon_images, mu, logvar = model(images)
        loss = vae_loss_function(recon_images, images, mu, logvar)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        progress_bar.set_postfix(loss=loss.item() / len(images))

    avg_loss = total_loss / len(dataset)
    loss_history.append(avg_loss)
    print(f"Epoch {epoch+1}/{NUM_EPOCHS} -- Average Loss: {avg_loss:.4f}")

    # Save reconstruction samples
    if (epoch + 1) % 10 == 0:
        with torch.no_grad():
            fixed_images = next(iter(dataloader)).to(device)
            recon_fixed, _, _ = model(fixed_images)
            comparison = torch.cat([fixed_images[:8], recon_fixed[:8]])
            save_image(comparison.cpu(), os.path.join(OUTPUT_DIR, f"reconstructions/recon_epoch_{epoch+1}.png"), nrow=8)

# --- 6. Save Final Model and Data ---
print("Training finished.")
torch.save(model.state_dict(), os.path.join(OUTPUT_DIR, "vae_model.pth"))

model.eval()
all_mu = []
with torch.no_grad():
    for images in dataloader:
        images = images.to(device)
        mu, _ = model.encode(images)
        all_mu.append(mu.cpu().numpy())
latent_vectors = np.concatenate(all_mu, axis=0)
np.save(os.path.join(OUTPUT_DIR, "latent_vectors.npy"), latent_vectors)

# Plot and save the loss curve
plt.figure(figsize=(10, 5))
plt.plot(loss_history)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.savefig(os.path.join(OUTPUT_DIR, "loss_curve.png"))
plt.show()

print("Model and latent vectors saved. You can now run the visualization cell.")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import umap
from torchvision.utils import make_grid

# We need to import the model definition again
from model import VAE

# --- 1. Parameters (should match training) ---
OUTPUT_DIR = 'outputs'
LATENT_DIM = 32
IMG_SIZE = 128

# --- 2. Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- 3. UMAP Visualization ---
print("Generating UMAP plot...")
latent_vectors = np.load(os.path.join(OUTPUT_DIR, "latent_vectors.npy"))
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, metric='euclidean')
embedding = reducer.fit_transform(latent_vectors)

plt.figure(figsize=(12, 10))
plt.scatter(embedding[:, 0], embedding[:, 1], s=5, alpha=0.7)
plt.title('UMAP Projection of the Latent Space')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.grid(True)
plt.savefig(os.path.join(OUTPUT_DIR, "umap_manifold.png"))
plt.show()

# --- 4. Generative Manifold Visualization ---
print("Generating 2D manifold...")
model = VAE(img_channels=1, latent_dim=LATENT_DIM).to(device)
model.load_state_dict(torch.load(os.path.join(OUTPUT_DIR, "vae_model.pth"), map_location=device))
model.eval()

grid_size = 20
z_samples = torch.randn(grid_size * grid_size, LATENT_DIM).to(device)
with torch.no_grad():
    generated_images = model.decode(z_samples).cpu()

grid = make_grid(generated_images, nrow=grid_size, padding=2, normalize=True)
plt.figure(figsize=(15, 15))
plt.imshow(grid.permute(1, 2, 0))
plt.title('Generated Manifold from Latent Space')
plt.axis('off')
plt.savefig(os.path.join(OUTPUT_DIR, "generative_manifold.png"))
plt.show()
```
